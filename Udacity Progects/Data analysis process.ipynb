# Load Data from CSVs

```python
import unicodecsv
```

```python
# информация зачислении и прекращении студентами курса

with open('enrollments.csv', 'rb') as f:
    reader = unicodecsv.DictReader(f)
    enrollments = list(reader)
```

```python
# информация о ежедневной активности студентов

with open('daily_engagement.csv', 'rb') as f:
    reader = unicodecsv.DictReader(f)
    daily_engagement = list(reader)
```

```python
# информация о представленных проектах 

with open('project_submissions.csv', 'rb') as f:
    reader = unicodecsv.DictReader(f)
    project_submissions = list(reader)
```

```python
from datetime import datetime as dt

def parse_date(date): # преобразует строковый тип в дату
    if date == '':
        return None
    else:
        return dt.strptime(date, '%Y-%m-%d')
    
def parse_int(i): # преобразует str в int
    if i == '':
        return None
    else:
        return int(i)
```

```python
# приводим к нужному типу данные в столбцах

for enrollment in enrollments:
    enrollment['join_date'] = parse_date(enrollment['join_date'])
    enrollment['cancel_date'] = parse_date(enrollment['cancel_date'])
    enrollment['days_to_cancel'] = parse_int(enrollment['days_to_cancel'])
    enrollment['is_udacity'] = enrollment['is_udacity'] == "True"
    enrollment['is_canceled'] = enrollment['is_canceled'] == "True"
    

for d in daily_engagement:
    d['utc_date'] = parse_date(d['utc_date'])
    d['num_courses_visited'] = int(float(d['num_courses_visited']))
    d['total_minutes_visited'] = float(d['total_minutes_visited'])
    d['lessons_completed'] = int(float(d['lessons_completed']))
    d['projects_completed'] = int(float(d['projects_completed']))
    
for i in project_submissions:
    i['creation_date'] = parse_date(i['creation_date'])
    i['completion_date'] = parse_date(i['completion_date'])
```

```python
# находит число уникальных студентов в датасете
def find_unique_students(data):
    num_unique_students = set()
    for i in data:
        num_unique_students.add(i['account_key'])
    return num_unique_students
```

```python
# нашли всколько всего студентов в датасетах и количество строк
enrollment_num_rows = len(enrollments)   
enrollment_num_unique_student = find_unique_students(enrollments)

engagement_num_rows =len(daily_engagement)
engagement_num_unique_studentt = find_unique_students(daily_engagement)

submission_num_rows =len(project_submissions)  
submission_num_unique_students = find_unique_students(project_submissions)
```

```python
# находим студентов которые учились не больше одного дня и 
# которые попали в число уникальных студентов в таблице активностей
# их не должно быть там, потому что это тестовые аккаунты
s = set()
for i in enrollments:
    if i['cancel_date'] != i['join_date'] and i['account_key'] not in engagement_num_unique_studentt:
        s.add(i['account_key'])
            
       
        
```

```python
# находим все тестовые аккаунты в списке enrollments

udacity_test_acc =set()
for i in enrollments:
    if i['is_udacity']:
        udacity_test_acc.add(i['account_key'])
len(udacity_test_acc)
```

```python
# функция, которая удаляет тестовые аккаунты 

def remove_udacity_acc(data):
    non_udacuty_acc = []
    for i in data:
        if i['account_key'] not in udacity_test_acc:
            non_udacuty_acc.append(i)
    return non_udacuty_acc
```

```python
# создаем новые списки без тестовых аккаунтов

non_udacity_enrollments = remove_udacity_acc(enrollments)
non_udacity_engagement = remove_udacity_acc(daily_engagement)
non_udacity_submissions = remove_udacity_acc(project_submissions)
```

```python
print(len(non_udacity_enrollments))
print(len(non_udacity_engagement))
print(len(non_udacity_submissions))
```

```python
paid_students = dict()
for i in  non_udacity_enrollments:
    if i['days_to_cancel'] == None or i['days_to_cancel'] > 7:
        acc_key = i['account_key']
        enrollment_date = i['join_date']
        
        if acc_key not in paid_students or enrollment_date > paid_students[acc_key]:
            paid_students[i['account_key']] = i['join_date']
        
```

```python
len(paid_students)
```

```python
def within_one_week(join_date,engagement_date):
    time_delta = enrollment_date - join_date
    return time_delta.days < 7
```

```python
paid_engagement_in_first_week = []
for i in non_udacity_engagement:
    if i['account_key'] in paid_students:
        if within_one_week(paid_students[i['account_key']], i['utc_date']):
            paid_engagement_in_first_week.append(i)
    
```
