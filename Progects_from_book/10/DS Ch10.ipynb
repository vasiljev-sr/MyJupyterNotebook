```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('AvarangeTempNY.csv')
```

```python
df.drop(['Anomaly'], axis=1,inplace=True)
```

```python
df.Date = df.Date.floordiv(100)
```

```python
df.columns = ["Date",'Temperature']
```

```python
df.head()
```

```python
linear_regression = stats.linregress(x=df.Date, y=df.Temperature)
x = df.Date
y = x*linear_regression.slope + linear_regression.intercept
plt.plot(x,y,'-')
plt.plot(df.Date,df.Temperature,'ro')
plt
```
